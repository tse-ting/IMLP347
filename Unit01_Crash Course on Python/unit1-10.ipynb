```python
fin_E = open("C:\\Users\\user\\Desktop\\python_machine\\unit01\\data\\english_list.csv", "r", encoding='UTF-8')
fin_M = open("C:\\Users\\user\\Desktop\\python_machine\\unit01\\data\\math_list.csv", "r", encoding='UTF-8')
lisE = []
lisM = []
name = []

# Read English scores
for line in fin_E.readlines():
    print(line, end='')
    line = line.strip().split(",")
    name.append(line[0])
    lisE.append(line[1])

# Read Math scores
for line in fin_M.readlines():
    print(line, end='')
    line = line.strip().split(",")
    lisM.append(line[1])

# Calculate and write total scores to CSV
fout = open("Score.csv", "w")
fout.write("Name,Total\n")  # Header

for i in range(1, len(name)):
    total_score = int(lisE[i]) + int(lisM[i])
    list1 = [name[i], str(total_score) + "\n"]
    print(name[i], total_score)
    line = ",".join(list1)
    fout.write(line)

# Close all opened files
fin_E.close()
fin_M.close()
fout.close()

```
